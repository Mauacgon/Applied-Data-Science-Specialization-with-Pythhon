```python
%matplotlib notebook

import matplotlib.pyplot as plt
import numpy as np
import matplotlib.gridspec as gridspec

fig = plt.figure()
gspec = gridspec.GridSpec(6,6)
top_left = plt.subplot(gspec[0:2,0:3])
top_right = plt.subplot(gspec[0:2,3:], sharey = top_left)
bottom_left = plt.subplot(gspec[3:5,0:3], sharey = top_left)
bottom_right = plt.subplot(gspec[3:5,3:], sharey = top_left)

x1 = np.random.normal(-2.5, 1, 10000)
x2 = np.random.gamma(2, 1.5, 10000)
x3 = np.random.exponential(2, 10000)+7
x4 = np.random.uniform(14,20, 10000)

top_left.hist(x1, bins = 100, density = True)
top_right.hist(x2, bins = 100, density = True)
bottom_left.hist(x3, bins = 100, density = True)
bottom_right.hist(x4, bins = 100, density = True)

top_left.set_title('Normal')
top_right.set_title('Gamma')
bottom_left.set_title('Exponential')
bottom_right.set_title('Uniform')

top_right.tick_params(axis = 'y', labelleft = 'off')
bottom_right.tick_params(axis = 'y', labelleft = 'off')
```
